```python
import math, copy
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
def load_data(filename):
    data = np.loadtxt(filename, delimiter=',')
    X = data[:, :2]
    y = data[:, 2]
    return X, y

def map_feature(X1, X2):
    """
    Feature mapping function to polynomial features
    """
    X1 = np.atleast_1d(X1)
    X2 = np.atleast_1d(X2)
    degree = 6
    out = []
    for i in range(1, degree+1):
        for j in range(i + 1):
            out.append((X1**(i-j) * (X2**j)))
    return np.stack(out, axis=1)


def plot_data(X, y, pos_label="y=1", neg_label="y=0"):
    positive = y == 1
    negative = y == 0
    
    # Plot examples
    plt.plot(X[positive, 0], X[positive, 1], 'k+', label=pos_label)
    plt.plot(X[negative, 0], X[negative, 1], 'yo', label=neg_label)

def plot_decision_boundary(w, b, X, y):
    # Credit to dibgerge on Github for this plotting code
     
    plot_data(X[:, 0:2], y)
    
    if X.shape[1] <= 2:
        plot_x = np.array([min(X[:, 0]), max(X[:, 0])])
        plot_y = (-1. / w[1]) * (w[0] * plot_x + b)
        
        plt.plot(plot_x, plot_y, c="b")
        
    else:
        u = np.linspace(-1, 1.5, 50)
        v = np.linspace(-1, 1.5, 50)
        
        z = np.zeros((len(u), len(v)))

        # Evaluate z = theta*x over the grid
        for i in range(len(u)):
            for j in range(len(v)):
                z[i, j] = sigmoid(np.dot(map_feature(u[i], v[j]), w) + b)
        
        # important to transpose z before calling contour
        z = z.T
        
        # Plot z = 0
        plt.contour(u, v, z, levels=[0.5], colors="g")
```

```python
# load dataset
plt.close('all')
X_train, y_train = load_data("data/ex2data2.txt")

# Plot examples
plot_data(X_train, y_train[:], pos_label="Accepted", neg_label="Rejected")

# Set the y-axis label
plt.ylabel('Microchip Test 2') 
# Set the x-axis label
plt.xlabel('Microchip Test 1') 
plt.legend(loc="upper right")
plt.show()
```

```python
epsilon = np.finfo('double').eps


def sigmoid(z):
    z = np.clip(z, -500, 500) # protect against overflow
    g = 1.0/(1.0+np.exp(-z))

    return g


def f_x(w, b, x):
    return sigmoid(np.dot(x, w) + b)


def compute_cost(w, b, x, y):
    m, n = x.shape
    total_cost = 0
    for i in range(m):
        y_hat = f_x(w, b, x[i])
        if y_hat < epsilon:
            y_hat = epsilon
        if 1 - y_hat < epsilon:
            y_hat = 1 - epsilon
        total_cost += (-y[i] * np.log(y_hat) - (1 - y[i]) * np.log(1 - y_hat))
    return total_cost / m


def gradient_descent(a, w, b, x, y):
    m, n = x.shape
    sum_dw = np.zeros((n,))
    sum_db = 0.
    
    w_in = copy.deepcopy(w)
    b_in = b
    
    for i in range(m):
        err_i = f_x(w, b, x[i]) - y[i]
        sum_dw += err_i * x[i]
        sum_db += err_i
    
    return w_in - a * sum_dw / m, b_in - a * sum_db / m
```

```python
print("Original shape of data:", X_train.shape)
mapped_X =  map_feature(X_train[:, 0], X_train[:, 1])
print("Shape after feature mapping:", mapped_X.shape)
```

```python
# UNQ_C5
def compute_cost_reg(w, b, X, y, lambda_ = 1):
    m, n = X.shape
    
    cost_without_reg = compute_cost(w, b, X, y) 
    
    reg_cost = 0.
    
    for i in range(n):
        reg_cost += w[i] * w[i]
    
    total_cost = cost_without_reg + (lambda_/(2 * m)) * reg_cost

    return total_cost

def gradient_descent_reg(a, w, b, X, y, lambda_ = 1): 
    m, n = X.shape
    sum_dw = np.zeros((n,))
    sum_db = 0.
    
    w_in = copy.deepcopy(w)
    b_in = b
    
    for i in range(m):
        err_i = f_x(w, b, X[i]) - y[i]
        sum_dw += err_i * X[i]
        sum_db += err_i
    
    return w_in - a * (sum_dw / m + (lambda_ / m) * w), b_in - a * sum_db / m

def predict(X, w, b): 
    """
    Predict whether the label is 0 or 1 using learned logistic
    regression parameters theta
    
    Parameters
    ----------
    X : array_like
        Shape (m, n+1) 
    
    w : array_like
        Parameters of the model
        Shape (n, 1)
    b : scalar
    
    Returns
    -------

    p: array_like
        Shape (m,)
        The predictions for X using a threshold at 0.5
        i.e. if sigmoid (theta.T*X) >=0.5 predict 1
    """
    
    # number of training examples
    m = X.shape[0]   
    p = np.zeros(m)
   
    for i in range(m):
        f_w = sigmoid(np.dot(w.T, X[i]) + b)
        p[i] = f_w >=0.5
    
    return p
```

```python
X_mapped = map_feature(X_train[:, 0], X_train[:, 1])

num_iters = 10000
np.random.seed(1)
w = np.random.rand(X_mapped.shape[1])-0.5
b = 1.

# Set regularization parameter lambda_ to 1 (you can try varying this)
lambda_ = 0.01;  

# Some gradient descent settings
alpha = 0.01

for i in range(num_iters):
    w, b = gradient_descent_reg(alpha, w, b, X_mapped, y_train, lambda_)   
        
    if i % (num_iters/10) == 0 or i == (num_iters -1):
        cost = compute_cost_reg(w, b, X_mapped, y_train, lambda_)
        print(f"Iteration={i}, cost={cost}")
```

```python
fig, ax = plt.subplots()
plot_decision_boundary(w, b, X_mapped, y_train)
```

```python
#Compute accuracy on the training set
p = predict(X_mapped, w, b)

print('Train Accuracy: %f'%(np.mean(p == y_train) * 100))
```
